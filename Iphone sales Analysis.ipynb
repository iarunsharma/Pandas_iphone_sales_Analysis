# Ques1:- Top 10 highest rated iPhone on Flipkart in india ?

```python
df.head()
```

```python
top10_rated_iPhones = df.groupby("Product Name")['Number Of Ratings'].sum().head(10)
top10_rated_iPhones
```

```python
sns.barplot(top10_rated_iPhones  , color = 'Chocolate')
plt.xticks(rotation=60, ha='right' , color = 'Peru')  # Rotate labels by 45 degrees and align them to the right
plt.show()
```

# Ques2:- How many Rating do the highest rated iPhone on flipkart have?

```python
# highest_rated_iphone=df[["Product Name","Number Of Ratings"]].head(5).sort_values('Number Of Ratings',ascending=False)
# highest_rated_iphone

highest_rated_iphone = df.groupby("Product Name")["Number Of Ratings"].sum().head(5)
highest_rated_iphone
```

```python
sns.barplot(highest_rated_iphone , color = 'FireBrick')
plt.xticks(rotation=60, ha='right' , color = 'red')  # Rotate labels by 45 degrees and align them to the right
plt.show()
```

# Ques 3:- Which iPhone has highest number of reviews on flipkart?

```python
df.head()
```

```python
# df[['Product Name','Number Of Reviews']].sort_values(by='Number Of Reviews', ascending=False).head(5)
highest_review = df.groupby('Product Name')['Number Of Reviews'].sum().head(5)
highest_review
```

# Ques4:- Can you figure out the least expensive and most expensive iphones in the Indian market, along with all their specification?

```python
df.head()
```

```python
most_exp=df.loc[df["Sale Price"].idxmax()]
least_exp=df.loc[df["Sale Price"].idxmin()]
print("Most expensive")
print(most_exp)
print("\nLeast expansive")
print(least_exp)
```
